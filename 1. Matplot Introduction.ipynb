# Matplot library introduction

### Histogram Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_list = np.random.randint(1, 100, size= 40)
random_list
```

```python
plt.hist(x=random_list)
#plt.xticks([2, 4, 8, 10, 20])
plt.show()
```

### Line Plot

```python
month = ['August', 'September', 'October', 'November', 'December']
students = [60, 70, 20, 34, 50]

plt.plot(month, students)
plt.show()
```

### Scatter Plot

```python
height = [5.5, 4.8, 5.5, 6, 5.4, 5.8]
weight = [60, 37, 70, 120, 47, 60]

plt.scatter(x=height, y=weight)
plt.show()
```

### Bar chart

```python
import pandas as pd
```

```python
raw_data = {'Name': ['upes', 'dit', 'gehu', 'dit', 'upes', 'dit', 'upes', 'gehu', 'dit', 'upes', 'upes', 'dit', 'dit', 'dit', 'gehu', 'upes', 'gehu']}

data = pd.DataFrame(data=raw_data)
data.Name.value_counts()
```

```python
font = {'family': 'serif',
        'weight': 'bold',
        'size': 20,
        }
plt.bar(x=data.Name.unique(), height=data.Name.value_counts())
plt.xlabel('College Names', fontdict=font)
plt.show()
```

### Pie Chart

```python
unique_colleges = data.Name.unique()
```

```python
unique_colleges
```

```python
sizes = data.Name.value_counts()
sizes
```

```python
plt.pie(sizes, labels=unique_colleges, autopct='%.2f', explode=(0, 0, 0.1))
plt.show()
```

### Box Plot

```python
box_data = np.random.randint(10, 50, size=50)
box_data
```

```python
outliers_data = np.random.randint(40, 150, size=10)
outliers_data
```

```python
new_data = np.concatenate((box_data, outliers_data), axis=0)
new_data
```

```python
plt.boxplot(new_data)
plt.show()
```
